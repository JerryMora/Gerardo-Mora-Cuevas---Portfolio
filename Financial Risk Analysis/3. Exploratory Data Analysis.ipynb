## Initial Visualizations

In order to make visualizations like graphs, is necessary to import the matplotlib and seaborn libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram for StrikePrice:

```python
sns.histplot(df["StrikePrice"], kde=True)
plt.show()
```

Histogram of LastPx:

```python
sns.histplot(df["LastPx"], kde=True)
plt.show()
```

```python
df[["StrikePrice","LastPx"]].corr()
```

The correlation between StrikePrice and LastPx is quite strong. That is why the histograms are very similar. Although, there are some slightly different values in both columns.

Let's create a scatter plot to explore relationships between StrikePrice and LastPx:

```python
sns.scatterplot(x="LastPx", y="StrikePrice",data=df, hue="Side")
plt.title("Scatter Plot: LastPx vs StrikePrice")
plt.show()
```

It seems to be some values in the LONG Side that difers from the rest.

Bar graph of the Symbol:

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Symbol",data=df,palette="viridis")
plt.title("Count of Symbol")
plt.xlabel("Symbol")
plt.ylabel("Count")
plt.show()
```

We can see now the count of each symbol used in the transactions. There are 13 unique symbols: "ESU0" and "NQU0" are the most frequent symbols, and "BTCK0" is the less frequent symbol.

```python
plt.figure(figsize=(4,6))
sns.countplot(x="Side", hue="CFICode", data=df)
plt.title("Count of Transactions by Side and CFICode")
plt.xlabel("Side")
plt.ylabel("Count")
plt.show()
```

This bargraph show the distribution of transactions by side in the only CFICode (MMMXXX). There are more transactions on the LONG side by far.

```python
df.head()
```

```python
df.set_index("TradeDate",inplace=True)
df.index = pd.to_datetime(df.index)
df_weekly = df.resample('W-Mon').size().reset_index(name="Number of Transactions")
plt.figure(figsize=(24,8))
sns.barplot(x=df_weekly["TradeDate"].dt.strftime('%Y-%m-%d'), y = df_weekly["Number of Transactions"],palette="Set2")
plt.title("Number of Transactions per week")
plt.xlabel("week start Date")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=90)
plt.show()
```

We can see the weekly distribution of transactions during 2020-05-18 to 2020-11-23.
